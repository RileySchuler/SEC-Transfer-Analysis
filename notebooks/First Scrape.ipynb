```python
loading_list = []
stat_index_dict = {'age': 0,
                           'team': 2,
                           'conference': 3,
                           'PA': 7,
                           'BB': 17,
                           'SO': 18,
                           'OBP': 20,
                           'SLG': 21,
                           'OPS': 22}
for player in soup.find_all('a', href=True)[484:504]: #find every url for 2025 MSST
    id = player['href']
    id_url = 'https://www.baseball-reference.com{0}'.format(id)
    temp_page = requests.get(id_url)
    time.sleep(1)
    temp_soup = BeautifulSoup(temp_page.text, 'html')
    for season in temp_soup.find_all('tr', class_="college full"): #find every college season for player
        
        temp_dict = {}
        temp_dict["player_id"] = id[25:] #obtains the last section of the url, which is the player_id  
        temp_dict["name"] = temp_soup.find('h1').get_text().strip()
        temp_dict["year"] = season.find_all('a')[0].get_text()
        for stat,index in stat_index_dict.items():
            temp_dict[stat] = season.find_all('td')[index].get_text()
        
        loading_list.append(temp_dict)
        print(temp_dict)
        time.sleep(1)
```
