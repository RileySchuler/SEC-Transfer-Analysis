```python
new_df = df.loc[df['player_id'] == df['player_id'].unique()[2]]
new_df
```

```python
def player_data_transformer(player_data):
    ll = []

    for i in range(len(player_data) - 1):
        prev_yr = player_data.iloc[i]
        next_yr = player_data.iloc[i+1]
        if prev_yr['year'] < 2021 or prev_yr['PA'] < 174 or next_yr['PA'] < 174 or next_yr['conference'] != 'SEC':
            continue
        
        temp_dict = {'player_id' : prev_yr['player_id'],
        'prev year' : prev_yr['year'],
        'next year' : next_yr['year'],
        'class transition' : str(i+1) + ' to ' + str(i+2)
        }

        if prev_yr['conference'] == 'SEC':
            temp_dict['type'] = 'Returning'
        else:
            temp_dict['type'] = 'Transfer'

        temp_dict['prev conf'] = prev_yr['conference']
        temp_dict['next conf'] = next_yr['conference']

        temp_dict['prev BB%'] = round(prev_yr['BB'] / prev_yr['PA'], 3) * 100
        temp_dict['next BB%'] = round(next_yr['BB'] / next_yr['PA'], 3) * 100

        temp_dict['prev K%'] = round(prev_yr['SO'] / prev_yr['PA'], 3) * 100
        temp_dict['next K%'] = round(next_yr['SO'] / next_yr['PA'], 3) * 100

        temp_dict['prev OBP'] = prev_yr['OBP']
        temp_dict['next OBP'] = next_yr['OBP']

        temp_dict['prev SLG'] = prev_yr['SLG']
        temp_dict['next SLG'] = next_yr['SLG']

        temp_dict['prev OPS'] = prev_yr['OPS']
        temp_dict['next OPS'] = next_yr['OPS']

        print(ll.append(temp_dict))
        
        
    return ll
```

```python
ll = []
print(player_data_transformer(new_df,ll))
```

```python
loading_list = []
for j in range(df['player_id'].nunique()):
    results = player_data_transformer(df.loc[df['player_id'] == df['player_id'].unique()[j]])
    loading_list.extend(results)
```

```python
loading_list
```

```python
transformed_df = pd.DataFrame(loading_list)
transformed_df
```

```python
trans_percent = (transformed_df['type'] == 'Transfer').mean() * 100
print(trans_percent)
print((trans_percent / 100) * len(transformed_df))
```
